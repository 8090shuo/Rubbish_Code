# Process USRP、data with & without zigbee signal

```python
# load library
import numpy as np
# from numpy import fft
import matplotlib.pyplot as plt 
import matplotlib.mlab as mlab
import scipy.signal
import scipy.interpolate
import scipy.ndimage
from sklearn.svm import SVC
import time
```

```python
def run_time(func):
    def wrapper(*args, **kw):
        start = time.process_time()
        res = func(*args, **kw)
        end = time.process_time()
        print("running time: %s seconds" % (end-start))
        return res
    return wrapper
```

```python
def cal_time(func):
    start = time.process_time()
    func(*args, **kw)
    end = time.process_time()
    print("running time: %s seconds" % (end-start))
```

```python
def load_zigbee_25M_data(n=1000):
    # load 1024*n IQdata
    IQdata = np.fromfile('usrp_25M_data_with_telosb.dat',dtype="float32",count=1024*2*n)
    # merge these data into complex form
    IQcomplex = map(complex,IQdata[0::2],IQdata[1::2])
    # change these data into complex type
    IQcomplex = np.array(list(IQcomplex),dtype=complex)
    # save data
#     IQdata[index*1024*2:(index+1)*1024*2].tofile("./usrp_81us.dat")
    return IQcomplex

def load_wifi_25M_data(n=100):
    # load 1024*n IQdata
    IQdata = np.fromfile('usrp_wifi_cf2433M_sr25M.dat',dtype="float32",count=1024*2*n)
    # merge these data into complex form
    IQcomplex = map(complex,IQdata[0::2],IQdata[1::2])
    # change these data into complex type
    IQcomplex = np.array(list(IQcomplex),dtype=complex)
    return IQcomplex

def load_data(filename='9.11usrp_zigbeeonwifi_rxgain10.dat' ,n=10000):
    # load 1024*n IQdata
    IQdata = np.fromfile(filename, dtype="float32", count=1024*2*n)
    # merge these data into complex form
    IQcomplex = map(complex,IQdata[0::2],IQdata[1::2])
    # change these data into complex type
    IQcomplex = np.array(list(IQcomplex),dtype=complex)
    
#     IQdata =  IQcomplex.real,IQcomplex.imag
    return IQcomplex

def load_low_power_data(filename='withtelosbpower_5dB_cf_2433M_SR_25M.dat' ,n=10000):
    # load 1024*n IQdata
    IQdata = np.fromfile(filename, dtype="float32", count=1024*2*n)
    # merge these data into complex form
    IQcomplex = map(complex,IQdata[0::2],IQdata[1::2])
    # change these data into complex type
    IQcomplex = np.array(list(IQcomplex),dtype=complex)
    return IQcomplex
```

```python

def cal_mag(IQcomplex,fft_size=1024):
#     fft_size = 1024
    IQcomplex_1024 = np.asarray(IQcomplex[0:fft_size])
    # 加窗，不然频谱现象不明显
    result, windowVal = mlab.apply_window(IQcomplex_1024,window=mlab.window_hanning,axis=0,return_window=True)
    # 进行fft
    result = np.fft.fft(result, n=fft_size)
    # 参数1：FFT点数，参数2：采样周期，即1/采样频率
    freqs = np.fft.fftfreq(fft_size, 1/25)
    
    # 以下代码的作用貌似是调整双边的位置
    freqcenter = fft_size//2
    freqs = np.concatenate((freqs[freqcenter:],freqs[:freqcenter]))
    result = np.concatenate((result[freqcenter:],result[:freqcenter]),0)
    
    # 貌似是归一化
    result = np.abs(result)/np.abs(windowVal).sum()
    # 加上中心频率
    Fc = 2433
    freqs += Fc 
    # 取对数坐标
    spec = 20.*np.log10(result)
    return spec,freqs
```

```python
# @run_time
def feature_cal(IQcomplex, n=1000):
    """
    find out whether it exists peak
    :param IQcomplex:
    :param n:
    :return: the center frequency of peaks and the indexs of the zigbee signal
    """
    center_freq = []
    bandwidth = []
    for index in range(n):
        IQcomplex_tmp = IQcomplex[index*1024:(index+1)*1024]
        mag, freqs = cal_mag(IQcomplex_tmp)
        mag_smooth = scipy.signal.savgol_filter(mag, 21, 3)
        indices = scipy.signal.find_peaks(mag_smooth, threshold=0.1, height=-60)[0]
        if len(indices) != 0:
            center_freq.extend([sum(freqs[indices]) / len(indices)])
            bandwidth.append(freqs[indices[-1]] - freqs[indices[0]])
        else:
            center_freq.extend([0])
            bandwidth.append(0)
    indexs = []
    for index, value in enumerate(center_freq):
        if value > 0:
            indexs.append(index)
    # print("索引："+ str(indexs))

    return center_freq, bandwidth, indexs
```

```python
def find_continuous_indexs(result):
    """
    this function is used for find out the continuous num of indexs
    :param indexs: the continuous list
    :return: the start and end element of each continuous list
    """
    indexs = []
    for index, value in enumerate(result):
        if value > 0:
            indexs.append(index)
    tnum = []
    tnum.append(indexs[0])
    for index in range(len(indexs) - 1):
        if indexs[index + 1] - indexs[index] == 1:
            continue
        else:
            tnum.append(indexs[index])
            tnum.append(0)
            tnum.append(indexs[index + 1])
    tnum.append(indexs[-1])
    data_index = map(lambda x,y: x-y, tnum[1::3], tnum[0::3])
    null_index = map(lambda x,y: x-y, tnum[3::3], tnum[1::3])
    print('the original continuous index are ' + str(list(tnum)))
    print('zigbee signal duration index is ' + str(list(data_index)))
    print('the gap betwwen zigbee signal is ' + str(list(null_index)))
    return tnum
```

```python
def plot_IQ(IQcomplex, IQIndex=0):
    """
    This function plot 10 figures, use IQIndex*100 us IQdata
    :param IQcomplex:
    :param IQIndex: the index num of IQcomplex
    :return:
    """
    plt.figure(figsize=(6, 10))
    num = IQIndex
    for n in range(10):
        plt.subplot(5, 2, n + 1)
        IQcomplex_tmp = IQcomplex[(n + num) * 1024:(n + 1 + num) * 1024]
        mag, freqs = cal_mag(IQcomplex_tmp)
        plt.plot(freqs, mag)
    plt.show()
    return

def plot_IQ_timeseries(IQcomplex, IQIndex=0):
    plt.figure(figsize=(6, 10))
    num = IQIndex
    for n in range(10):
        plt.subplot(5, 2, n + 1)
            # for time analysis
        bins = 1024
        IQcomplex_tmp_time = np.linspace(0, bins/25, bins)
        IQcomplex_tmp = IQcomplex[(n + num) * bins:(n + 1 + num) * bins]
        #plot time IQ seperate
#         plt.plot(IQcomplex_tmp_time,IQcomplex_tmp.real)
#         plt.plot(IQcomplex_tmp_time,IQcomplex_tmp.imag)
        
        # plot energy
#         IQcomplex_tmp_time = np.linspace(0, bins*2/25, bins)
        mag = np.sqrt(IQcomplex_tmp.real**2+IQcomplex_tmp.imag**2)
        plt.plot(IQcomplex_tmp_time, mag)
        plt.xlabel('Time(us)')
    plt.show()
    return

def plot_IQ_time_spectrum(IQcomplex, SpecificIndex):
    n = SpecificIndex
    plt.subplot(121)
    IQcomplex_tmp = IQcomplex_wifi_with_zigbee[(n) * 1024:(n + 1) * 1024]
    mag, freqs = cal_mag(IQcomplex_tmp)
    plt.plot(freqs, mag)
    plt.xlabel('Freq(MHz)')

    plt.subplot(122)
    bins = 1024
    IQcomplex_tmp_time = np.linspace(0, bins/25, bins)
    mag = np.sqrt(IQcomplex_tmp.real**2+IQcomplex_tmp.imag**2)
    plt.plot(IQcomplex_tmp_time, mag)
    plt.xlabel('Time(us)')
    plt.tight_layout()
    return
```

```python
## load zigbee data
IQcomplex_zigbee = load_zigbee_25M_data(n=1000)

## get feature
center_freqs_zigbee, bandwidths_zigbee, indexs_zigbee = feature_cal(IQcomplex_zigbee, n=1000)

labels_zigbee = np.zeros(len(center_freqs_zigbee))

for index in indexs_zigbee:
    labels_zigbee[index] = 1
labels_zigbee = np.array(labels_zigbee)

## load wifi data and get its features
IQcomplex_wifi = load_wifi_25M_data(n=1000)
center_freqs_wifi, bandwidths_wifi, indexs_wifi = feature_cal(IQcomplex_wifi, n=1000)
labels_wifi = np.zeros(len(center_freqs_wifi))

## get its features
feature_center_freqs = np.concatenate((center_freqs_zigbee,center_freqs_wifi))
feature_bandwidth = np.concatenate((bandwidths_zigbee,bandwidths_wifi))
labels = np.concatenate((labels_zigbee,labels_wifi))

feature = np.array([feature_center_freqs,feature_bandwidth])
print(feature.T.shape,labels.shape)
```

```python
clf = SVC(kernel='linear')
clf.fit(feature.T, labels)
```

```python
feature_wifi = np.array([center_freqs_wifi,bandwidths_wifi])
feature_zigbee = np.array([center_freqs_zigbee,bandwidths_zigbee])
result = clf.predict(feature_zigbee.T)
sum(labels-result)
```
